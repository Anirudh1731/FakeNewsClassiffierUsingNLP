```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
message=X.copy()
```

```python
message['title'][1]
```

```python
import nltk

import re

from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# message['title'][6]
# message.shape
message['title'].head(10)

# to make the index same order as initially it is in different order
message.reset_index(inplace=True)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(message)):

  review=re.sub('[^a-zA-Z]',' ',message['title'][i])
  review=review.lower()
  review=review.split()

  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)

  corpus.append(review)
```

```python
corpus
```

```python
voc_size=5000
one_hot_rep=[one_hot(word,voc_size) for word in corpus]
```

```python
one_hot_rep
```

```python
# adding padding to make all sentances of equal size or else we cannot train to embedding layer
sent_length=20
# all sentances are below 20 so we can keep this as max length of title
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
```

```python
embedded_docs[0]

# by adding padding we have made each sentance of length 20
```

```python
#model creation
embedding_vector_features=40
model=Sequential()

model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))

model.add(LSTM(100))

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
import numpy as np

x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_teat=train_test_split(x_final,y_final,test_size=0.33,random_state=42)
```

```python
#training the model

model.fit(x_train,y_train,epochs=20,batch_size=64)
```

```python
#this is overfit model so will add dropout

from tensorflow.keras.layers import Dropout
embedding_vector_features=40
model=Sequential()

model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=64)
```

```python
y_pred=model.predict(x_test)

y_pred=np.where(y_pred>0.5 ,1,0)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_teat,y_pred))
```
